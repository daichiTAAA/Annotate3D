# システムの目的
## 背景
### 組立作業における品質管理
組立作業において、品質管理は非常に重要である。そのため、不良の発生防止と、不良の早期発見・早期対応が求められる。
### AIの活用
AIの活用により、不良の発生防止と、不良の早期発見・早期対応を行うことができる。
### AI活用における課題
AI活用において次の行程が必要である。
- AIの学習に必要なデータの収集
- AIの学習に必要なデータのラベリング  

しかし、この工程において次の課題がある。
- データの収集には、多くの人手が必要である。
- データのラベリングには、多くの人手が必要である。
- データのラベリングには、専門知識が必要である。
- データのラベリングには、多くの時間が必要である。
- データのラベリングには、多くのコストが必要である。
## 目的
AI活用における課題を解決するために、次の目的を設定する。
- データの収集に必要な人手を削減する。
- データのラベリングに必要な人手を削減する。
- データのラベリングに必要な専門知識を削減する。
- データのラベリングに必要な時間を削減する。
- データのラベリングに必要なコストを削減する。

# 業務要件
* リアルタイムで不良検出を行うことで不良の流出を防ぐ
* リアルタイムで作業指示を行うことで作業ミスを防ぐ

# 機能要件
データの収集とラベリングに必要なコストを削減するために、次の機能を実装する。
* 3D CADデータを点群データに変換する
* 3D CAD点群データで点群物体検知モデルを学習する
* 2つ以上のカメラから撮影した物体の点群形状を復元する
* 点群物体検知モデルで物体検知を行う
* 物体検知した画像に3D CADのメタデータを使って品番や品名をアノテーションする
